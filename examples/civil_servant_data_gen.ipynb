```python
import random

# 定义姓氏列表
last_names = ['赵', '钱', '孙', '李', '周', '吴', '郑', '王', '冯', '陈', '褚', '卫', '蒋', '沈', '韩', '杨', '朱', '秦', '尤', '许',  
              '何', '吕', '施', '张', '孔', '曹', '严', '华', '金', '魏', '陶', '姜', '戚', '谢', '邹', '喻', '柏', '水', '窦', '章',  
              '云', '苏', '潘', '葛', '奚', '范', '彭', '郎', '鲁', '韦', '昌', '马', '苗', '凤', '花', '方', '俞', '任', '袁', '柳',  
              '酆', '鲍', '史', '唐', '费', '廉', '岑', '薛', '雷', '贺', '倪', '汤', '滕', '殷', '罗', '毕', '郝', '邬', '安', '常',  
              '乐', '于', '时', '傅', '皮', '卞', '齐', '康', '伍', '余', '元', '卜', '顾', '孟', '平', '黄', '和', '穆', '萧', '尹',  
              '姚', '邵', '湛', '汪', '祁', '毛', '禹', '狄', '米', '贝', '明', '臧', '计', '伏', '成', '戴', '谈', '宋', '茅', '庞',  
              '熊', '纪', '舒', '屈', '项', '祝', '董', '梁']
second_names = ['的', '一', '是', '在', '不', '了', '人', '有', '这', '个', '上', '大', '来', '到', '说', '生',
'国', '和', '地', '为', '子', '中', '你', '我', '以', '他', '时', '要', '她', '出', '就', '都',
'可', '里', '后', '那么', '会', '对', '学', '着', '能', '时', '那', '得', '起', '看', '天', '么',
'好', '去', '故', '想', '把', '使', '道', '老', '因', '开', '还', '起来', '总', '样', '见', '走',
'很', '说话', '本', '只', '下', '听', '真', '打', '再', '做', '把']
# 定义名词列表  
first_names = ['健','军','伟','刚', '勇','毅','俊','峰','强','军','杰','健']

def genNames(num):
	# 生成不重复随机人名 
	names = set()
	while len(names) < num:
			name = random.choice(last_names) + random.choice(second_names) + random.choice(first_names) 
			names.add(name)
	return names

names = genNames(200)
print(list(names))
print(len(list(names)))
```

生成随机时间

将时间转换为date格式

```python
import numpy as np
import random

def getRandomDate(year_begin, year_end):
    # 生成随机年份
    random_year = random.randint(year_begin, year_end)

    # 生成随机月份 
    random_month = random.randint(1, 12)

    # 生成随机天数
    # 闰年2月29天,平年2月28天  
    if random_month == 2:
        if ((random_year % 4 == 0) and (random_year % 100 != 0)) or (random_year % 400 == 0):
            random_day = random.randint(1, 29)
        else:
            random_day = random.randint(1, 28)
    elif random_month in [1, 3, 5, 7, 8, 10, 12]:
        random_day = random.randint(1, 31)
    else:
        random_day = random.randint(1, 30) 

    return np.datetime64(f'{random_year}-{random_month:02}-{random_day:02}')

def formatDate(date):
    return np.datetime_as_string(date, unit='M').replace('-', '.')

print(getRandomDate(1963, 2000))
```

```python

def genEmployee(jobs, members, names):
	depts = ['部门一','部门二', '部门三', '部门四', '部门五']
	jobIds = list(range(1, len(jobs)))
	jobWeights = list(reversed(jobIds)) # 权重是id的倒置
	jobWeightResult = random.choices(jobIds, weights=jobWeights, k=len(names))
	sex = ['男', '女']

	for idx, name in enumerate(names): 
			birth_date = getRandomDate(1955, 2000)
			entry_date = getRandomDate(birth_date.astype(datetime.date).year+20, 2020)
			last_promotion_date = getRandomDate(entry_date.astype(datetime.date).year+2, 2022)
			rec = {
					'姓名': name,
					'现工作单位及职务': depts[random.randint(0, 4)]+'\\n'+jobs[jobWeightResult[idx]],
					'性别': sex[random.randint(0, 1)],
					'民族': '汉',
					'籍贯': '四川成都',
					'出生年月': formatDate(birth_date),
					'参加工作时间': formatDate(entry_date),
					'任现职务时间': formatDate(last_promotion_date),
					'任现职级时间': formatDate(last_promotion_date),
					'入党时间': '1997.02',
					'全日制学历学位': '大学法学学生',
					'全日制毕业院校及专业': 'xx学院xx专业',
					'在职学历学位': '大学',
					'在职毕业院校及专业': 'xx大学xx专业',
					'任二主时间': '',
			}
			members.append(rec)
# 定义人员表格数据
members = []
jobs1 = ['一级调研员','二级调研员','三级调研员','四级调研员','一级主任科员','二级主任科员','三级主任科员','四级主任科员','一级科员','二级科员', '副处长' ,'处长']
jobs2 = ["二级行政执法员","一级行政执法员","四级主办","三级主办","二级主办","一级主办","四级高级主办","三级高级主办","二级高级主办","一级高级主办","副支队长","支队长"]
genEmployee(jobs=jobs1, members=members, names = genNames(300))
genEmployee(jobs=jobs2, members=members, names = genNames(350))
```
