生成随机时间

将时间转换为date格式

```python
import numpy as np
import random

def getRandomDate(year_begin, year_end):
    # 生成随机年份
    random_year = random.randint(year_begin, year_end)

    # 生成随机月份 
    random_month = random.randint(1, 12)

    # 生成随机天数
    # 闰年2月29天,平年2月28天  
    if random_month == 2:
        if ((random_year % 4 == 0) and (random_year % 100 != 0)) or (random_year % 400 == 0):
            random_day = random.randint(1, 29)
        else:
            random_day = random.randint(1, 28)
    elif random_month in [1, 3, 5, 7, 8, 10, 12]:
        random_day = random.randint(1, 31)
    else:
        random_day = random.randint(1, 30) 

    return np.datetime64(f'{random_year}-{random_month:02}-{random_day:02}')

def formatDate(date):
    return np.datetime_as_string(date, unit='M').replace('-', '.')

print(getRandomDate(1963, 2000))
```

```python
# 定义人员表格数据
jobs = ['一级调研员','二级调研员','三级调研员','四级调研员','一级主任科员','二级主任科员','三级主任科员','四级主任科员','一级科员','二级科员', '副处长' ,'处长']
jobIds = list(range(1, len(jobs)))
jobWeights = list(reversed(jobIds)) # 权重是id的倒置
jobWeightResult = random.choices(jobIds, weights=jobWeights, k=len(names))
sex = ['男', '女']
members = []
for idx, name in enumerate(names): 
    birth_date = getRandomDate(1955, 2000)
    entry_date = getRandomDate(birth_date.astype(datetime.date).year+20, 2020)
    last_promotion_date = getRandomDate(entry_date.astype(datetime.date).year+2, 2022)
    rec = {
        '姓名': name,
        '现工作单位及职务': jobs[jobWeightResult[idx]],
        '性别': sex[random.randint(0, 1)],
        '民族': '汉',
				'籍贯': '四川成都',
				'出生年月': formatDate(birth_date),
				'参加工作时间': formatDate(entry_date),
				'任现职务时间': formatDate(last_promotion_date),
				'任现职级时间': formatDate(last_promotion_date),
				'入党时间': '1997.02',
				'全日制学历学位': '大学法学学生',
				'全日制毕业院校及专业': 'xx学院xx专业',
				'在职学历学位': '大学',
				'在职毕业院校及专业': 'xx大学xx专业',
				'任二主时间': '',
    }
    members.append(rec)
```

```python
# 将字典转化为DataFrame
df = pd.DataFrame(members)
df.groupby('现工作单位及职务').size().head(100)
# df.head(100)
```
