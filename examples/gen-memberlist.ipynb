## 生成随机人名

```python
import random

# 定义姓氏列表
last_names = ['赵', '钱', '孙', '李', '周', '吴', '郑', '王', '冯', '陈', '褚', '卫', '蒋', '沈', '韩', '杨', '朱', '秦', '尤', '许',  
              '何', '吕', '施', '张', '孔', '曹', '严', '华', '金', '魏', '陶', '姜', '戚', '谢', '邹', '喻', '柏', '水', '窦', '章',  
              '云', '苏', '潘', '葛', '奚', '范', '彭', '郎', '鲁', '韦', '昌', '马', '苗', '凤', '花', '方', '俞', '任', '袁', '柳',  
              '酆', '鲍', '史', '唐', '费', '廉', '岑', '薛', '雷', '贺', '倪', '汤', '滕', '殷', '罗', '毕', '郝', '邬', '安', '常',  
              '乐', '于', '时', '傅', '皮', '卞', '齐', '康', '伍', '余', '元', '卜', '顾', '孟', '平', '黄', '和', '穆', '萧', '尹',  
              '姚', '邵', '湛', '汪', '祁', '毛', '禹', '狄', '米', '贝', '明', '臧', '计', '伏', '成', '戴', '谈', '宋', '茅', '庞',  
              '熊', '纪', '舒', '屈', '项', '祝', '董', '梁']

# 定义名词列表  
first_names = ['健','军','伟','刚', '勇','毅','俊','峰','强','军','杰','健']

# 生成不重复随机人名 
names = set()
while len(names) < 500:
    name = random.choice(last_names) + random.choice(first_names)
    names.add(name)

print(list(names))
```

生成随机时间

将时间转换为date格式

```python
import numpy as np
import random

def getRandomDate(year_begin, year_end):
    # 生成随机年份
    random_year = random.randint(year_begin, year_end)

    # 生成随机月份 
    random_month = random.randint(1, 12)

    # 生成随机天数
    # 闰年2月29天,平年2月28天  
    if random_month == 2:
        if ((random_year % 4 == 0) and (random_year % 100 != 0)) or (random_year % 400 == 0):
            random_day = random.randint(1, 29)
        else:
            random_day = random.randint(1, 28)
    elif random_month in [1, 3, 5, 7, 8, 10, 12]:
        random_day = random.randint(1, 31)
    else:
        random_day = random.randint(1, 30) 

    return np.datetime64(f'{random_year}-{random_month:02}-{random_day:02}')

print(getRandomDate(1963, 2000))
```

```python
# 定义人员表格数据
jobs = ['一级科员', '二级科员', '三级科员','四级科员', '副处长' ,'处长']
members = []
for name in names: 
    jobId = random.randint(0, 5)
    birth_date = getRandomDate(1955, 2000)
    entry_date = getRandomDate(birth_date.astype(datetime.date).year+20, 2020)
    last_promotion_date = getRandomDate(entry_date.astype(datetime.date).year+2, 2022)
    rec = {
        'name': name,
        'job': jobs[jobId],
        'birth_date': birth_date,
        'entry_date': entry_date,
        'last_promotion_date': last_promotion_date,
    }
    members.append(rec)
```

```python
# 将字典转化为DataFrame
df = pd.DataFrame(members)
df.head()
```
