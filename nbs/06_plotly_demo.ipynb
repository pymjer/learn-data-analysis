## 基本图形
参见: [plotly文档](https://plotly.com/python/text-and-annotations/)

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000,4), index=ts.index, columns=list("ABCD"))
df = df.cumsum()
plt.figure()
df.plot()
plt.legend(loc='best')
```

## 热力图

```python
import plotly.figure_factory as ff

z = [[1.1, .3, .5, .7, .9],
     [1, .8, .6, .4, .2],
     [.2, 0, .5, .7, .9],
     [.9, .8, .4, .2, 0]]

fig = ff.create_annotated_heatmap(z)
fig.show()
```

### 热力图自定义颜色

```python
import plotly.figure_factory as ff

z = [[1.1, .3, .5, .7, .9],
     [1, .8, .6, .4, .2],
     [.2, 0, .5, .7, .9],
     [.9, .8, .4, .2, 0]]

# 添加参数：colorscale
fig = ff.create_annotated_heatmap(z,colorscale='Viridis')

fig.show()
```

## 矩形图
柱形图有两种histogram和bar

- hostogram用来描述numerical变量，需要指定参数nbins，表示变量的长度
- bar用来描述categorical类型的变量，对变量进行分类

```python
import plotly.express as px
df = px.data.tips()
fig = px.histogram(df, x="total_bill", nbins=5)
fig.show()
```

```python
df
```

```python
px.bar(df, x=df.index, y="total_bill")
```

调整axis

```python
import plotly.express as px
fig = px.bar(x=["a","a","b",3], y=[1,2,3,4])
fig.show()
```

```python
fig.update_xaxes(type='category')
fig.show()
```

## 散列图(Scatter)

```python
import plotly.express as px
df = px.data.iris()
print(df)
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                 size='petal_length', hover_data=['petal_width'])
fig.show()
```

### 散列图配置边缘分布图

```python
import plotly.express as px
df = px.data.iris()
fig = px.scatter(df, x="sepal_length", y="sepal_width",  marginal_x = "box",marginal_y = "violin")
fig.show()
```
